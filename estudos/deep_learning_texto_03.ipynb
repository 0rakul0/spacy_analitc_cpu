```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Embedding
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
dataset = pd.read_csv('../dados/spam/spam.csv')
dataset.head()
```

# tratamento de dados
### aqui é um problema de classificação se é ou não SPAM
### transforma em numeros

### transforma as respostas

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(dataset['Category'])
```

### pega os textos

```python
mensagens = dataset['Message'].values
```

### separa em treino e teste

```python
X_traino, X_teste, y_treino, y_teste = train_test_split(mensagens, y, test_size=0.3)
```

### embedding dos textos

```python
token = Tokenizer(num_words=100)
token.fit_on_texts(X_traino)
```

```python
X_traino = token.texts_to_sequences(X_traino)
X_teste = token.texts_to_sequences(X_teste)
```

### O que é feito aqui é um seguinte, é uma lista que quando a palavra aparece no indice ele marca como 1

```python
print(X_traino)
```

### com este metodo precisamos dá a todas as linhas o mesmo tamanho usando o pad

```python
tamanho_maximo = 500
```

```python
X_traino = pad_sequences(X_traino, padding="post", maxlen=tamanho_maximo) # coloca a direta [2,5,8,2,0,0,0,0,0]
X_teste = pad_sequences(X_teste, padding="post", maxlen=tamanho_maximo) # coloca a direta [2,5,8,2,0,0,0,0,0]
```

```python
print(X_traino)
```

### para ver a quantidade de palavras diferentes

```python
print(len(token.word_index))
```

### iniciando o modelo

```python
modelo = Sequential()
```

### adicionando uma camada embadding

```python
modelo.add(Embedding(input_dim=len(token.word_index), output_dim=50, input_length=tamanho_maximo))
```

```python
modelo.add(Flatten())
```

### camadas densas

```python
modelo.add(Dense(units=10, activation="relu"))
modelo.add(Dropout(0.1))
modelo.add(Dense(units=1, activation="sigmoid"))
```

### após o modelo criado precisamos compilar o modelo

```python
modelo.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
```

```python
modelo.summary()
```

treinando o modelo

```python
modelo.fit(X_traino, y_treino, epochs=40, batch_size=15, verbose=True, validation_data=(X_teste, y_teste))
```

#### para avaliação do modelo

```python
loss, accuracy = modelo.evaluate(X_teste, y_teste)
print(f"loss: {loss}\n accuracy: {accuracy}")
```

```python
nova_previsao = modelo.predict(X_teste)
e_spam = (nova_previsao>0.7)
print(e_spam)
```

### matriz de confusao

```python
cm = confusion_matrix(y_teste, e_spam)
print(cm)
```

```python
modelo.save("../modelo.h5")
```

- input_dim = Tamanho de entradas do vocabulo
- output_dim = Tamanho do vetor denso
- input_length = tamanho maximo da sequência
