```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
dataset = pd.read_csv('../dados/spam/spam.csv')
dataset.head()
```

# tratamento de dados
### aqui é um problema de classificação se é ou não SPAM
### transforma em numeros

### transforma as respostas

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(dataset['Category'])
```

### pega os textos

```python
mensagens = dataset['Message'].values
```

### separa em treino e teste

```python
X_traino, X_teste, y_treino, y_teste = train_test_split(mensagens, y, test_size=0.3)
```

### vetorizar as menssagens

```python
vetorizador = CountVectorizer()
vetorizador.fit(X_traino)
X_traino = vetorizador.transform(X_traino)
X_teste = vetorizador.transform(X_teste)
print(X_traino.toarray())
```

### O que é feito aqui é um seguinte, é uma lista que quando a palavra aparece no indice ele marca como 1

```python
import numpy as np
```

```python
np.set_printoptions(threshold=np.inf)
X_traino.toarray()[5]
```

### iniciando o modelo

```python
modelo = Sequential()
```

### camadas densas

#### para textos presisamos de um input_dim dinamico, logo quando vetorizamos podemos passar o tamanho do nosso vetor de texto

```python
input_camada = X_traino.shape[1]
```

```python
modelo.add(Dense(units=10, activation="relu", input_dim=input_camada))
modelo.add(Dropout(0.1))
modelo.add(Dense(units=8, activation="relu"))
modelo.add(Dropout(0.1))
modelo.add(Dense(units=1, activation="sigmoid"))
```

### após o modelo criado precisamos compilar o modelo

```python
modelo.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
```

```python
modelo.summary()
```

treinando o modelo

```python
modelo.fit(X_traino, y_treino, epochs=40, batch_size=15, verbose=True, validation_data=(X_teste, y_teste))
```

#### para avaliação do modelo

```python
loss, accuracy = modelo.evaluate(X_teste, y_teste)
print(f"loss: {loss}\n accuracy: {accuracy}")
```

```python
nova_previsao = modelo.predict(X_teste)
e_spam = (nova_previsao>0.7)
print(e_spam)
```

### matriz de confusao

```python
cm = confusion_matrix(y_teste, e_spam)
print(cm)
```

```python
modelo.save("../modelo.h5")
```
